```python
import traceback
for tuple in degrees:
    degree_1, degree_2 = tuple[0], tuple[1]
    if degree_2 is not None:
        print('----------------------- ' + degree_1 + ' ----------------------- ' + degree_2 + ' ---------------------------')
    else:
        print('----------------------- ' + degree_1 + ' ----------------------- ')
    try: 
        irt_analysis(degree_1=degree_1, 
                    degree_2=degree_2,
                    pass_grades=pass_grades,
                    fail_grades=fail_grades)
    except Exception as e:
        print('Error in irt_analysis. Error message: ' + str(e))
        print(traceback.format_exc())
        continue
```
